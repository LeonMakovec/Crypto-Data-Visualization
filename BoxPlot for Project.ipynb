```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

crypto = pd.read_csv('consolidated_coin_data.csv')

bc_eth = crypto[crypto['Currency'].isin(['bitcoin', 'ethereum'])]

# Convert 'Close' column to numeric type
bc_eth.loc[:, 'Close'] = pd.to_numeric(bc_eth['Close'], errors='coerce')

# Remove commas from 'Volume' column and convert to numeric type
bc_eth.loc[:, 'Volume'] = bc_eth['Volume'].str.replace(',', '')
bc_eth.loc[:, 'Volume'] = pd.to_numeric(bc_eth['Volume'], errors='coerce')

#Create for Close
plt.figure(figsize=(10, 6))
sns.boxplot(data=bc_eth, x='Currency', y='Close', palette='Set2')
plt.title('Comparison of Closing Price for Bitcoin and Ethereum')
plt.xlabel('Currency')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)

plt.legend(labels=['Bitcoin', 'Ethereum'], title='Currency')

plt.tight_layout()
plt.show()

#Create for Volume 
plt.figure(figsize=(10, 6))
sns.boxplot(data=bc_eth, x='Currency', y='Volume', palette='Set2')
plt.title('Comparison of Trading Volume for Bitcoin and Ethereum')
plt.xlabel('Currency')
plt.ylabel('Trading Volume')
plt.xticks(rotation=45)

plt.ylim(0, 1.17e10)

plt.legend(labels=['Bitcoin', 'Ethereum'], title='Currency')

plt.tight_layout()
plt.show()
```
